```python
def fix_vital_status(dataframe,list_of_unique_case_ids):
    for caseid in list_of_unique_case_ids:
        subset = dataframe[dataframe.bcr_patient_uuid == caseid]
        vital_value = subset.vital_status.dropna().values[0]
        subset.vital_status = subset.vital_status.map({np.nan:vital_value, vital_value:vital_value})
        dataframe[dataframe.bcr_patient_uuid == caseid] = subset
    return dataframe
       
```

```python
clean_cdf = fix_vital_status(clinical_data, clinical_data.bcr_patient_uuid.value_counts().index)
```
